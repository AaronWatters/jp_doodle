# Bounded Value Slider

<img src="bounded_value_slider.png" width="500"/>

This notebook demonstrates how to use `bounded_value_sliders` which identify
three values in a specified range -- the low boundary, the high boundary and the current value.

Boundary value sliders were introduced initially to support the `volume` widget
in the <a href="https://github.com/AaronWatters/feedWebGL2">feedWebGL2</a> package.

They can be used within a proxy widget interface or within vanilla Javascript as a jQuery plugin.

The cells below demonstrate the use of the proxy widget interface.  Drag the colored circles
to change the selected values in the widget.

```python
from jp_doodle import bounded_value_slider
```

```python
def on_stop(D):
    "Callback that uses the values generated by the bounded value widget."
    print("got values", D)
    
# Horizontal slider (default)
w = bounded_value_slider.BoundedValueSlider(
    on_stop=on_stop, 
    integral=True, 
    aspect_ratio=0.1,
    forbidden="red",
)

w.debugging_display()
```

```python
# Vertical slider (horizontal=False)
w2 = bounded_value_slider.BoundedValueSlider(
    on_stop=on_stop, 
    integral=False, 
    horizontal=False,
    aspect_ratio=0.08,
    length=333,
)

w2.debugging_display()
```
